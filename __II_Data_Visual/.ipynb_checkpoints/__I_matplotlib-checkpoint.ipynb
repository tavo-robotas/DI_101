### Nustatymai

Duokime savo grafikui pavadinimą:

```python
plt.title('Lietuvos vidutinis atlyginimas (€)')
plt.plot(periods, neto);
```

Pridėkime pavadinimus mūsų x ir y ašims.

```python
plt.xlabel('Metai')
plt.ylabel('Neto atlyginimas €')
plt.title('Lietuvos vidutinis atlyginimas (€)')
plt.plot(periods, neto);
```

Pridėkime prie mūsų grafiko papildomos informacijos, papildomai prie neto pridėkime bruto.

```python
plt.xlabel('Metai')
plt.ylabel('Atlyginimas €')
plt.title('Lietuvos vidutinis atlyginimas (€)')
plt.plot(periods, neto)
plt.plot(periods, bruto);
```

Dabar turime dvi kreives. But mes nežinome kuri yra kuri. Todel turėsime pridėti paiškinimų legendą, tai padaroma yra persiunčiam `plt.plot()` metodui dar vieną argumentą `label` su grafiko pavadinimu kuris bus nupieštas jeigu pakviesime metodą `plt.legend()`.

```python
plt.xlabel('Metai')
plt.ylabel('Atlyginimas €')
plt.title('Lietuvos vidutinis atlyginimas (€)')
plt.plot(periods, neto, label='Neto')
plt.plot(periods, bruto,  label='Bruto')
plt.legend();
```

Mes taip pat galime kontroliuoti savo kreiviu spalvas ir kitas savybes ir tą galima daryto dviem būdais. Vienas iš jų naudojant `format string` : https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

```python
fmt = '[marker][line][color]'
```

Tarkime mes savo grafike norime geltonos, punktirinės linijos.

```python
plt.xlabel('Metai')
plt.ylabel('Atlyginimas €')
plt.title('Lietuvos vidutinis atlyginimas (€)')
plt.plot(periods, neto, 'y--',label='Neto')
plt.plot(periods, bruto, 'b',  label='Bruto')
plt.legend();
```

Bet yra antras, aiškesnis ir mano manymu patogesnis būdas kontroliuoti kreivių išvaizdą.

```python
plt.xlabel('Metai')
plt.ylabel('Atlyginimas €')
plt.title('Lietuvos vidutinis atlyginimas (€)')
plt.plot(periods, neto, color='y', linestyle='--', label='Neto')
plt.plot(periods, bruto,  color='b', label='Bruto')
plt.legend();
```

Galima papildyte savo savo kreives markeriais.

```python
plt.xlabel('Metai')
plt.ylabel('Atlyginimas €')
plt.title('Lietuvos vidutinis atlyginimas (€)')
plt.plot(periods, neto, color='y', linestyle='--', marker='.', label='Neto')
plt.plot(periods, bruto,  color='b', marker='o', label='Bruto')
plt.legend();
```

Kreivių spalvas tai pat galima kontroliuoti su `hex` reikšmėmis `color` argumente. Kas yra daug paprasčiau ir aiškaiu nei prieš tai demonstruotas būdas. Norint išsrinkti spalvą tai galima padaryti su google color picker : https://g.co/kgs/MB68Nu

```python
plt.xlabel('Metai')
plt.ylabel('Atlyginimas €')
plt.title('Lietuvos vidutinis atlyginimas (€)')
plt.plot(periods, neto, color='#fff81f', linestyle='--', marker='.', label='Neto')
plt.plot(periods, bruto,  color='b', marker='o', label='Bruto')
plt.legend();
```

Įtraukime į grafika papildomos informacijos. Lietuvos BVP vienam gyventojui tuo pačiu periodu nuo 2008 iki 2021. Tai pat pridėsi papildomu argumentų šios kreivės `plt.plot` metodui, tokių kaip linijos storis `linewidth`.

```python
gdp_per_capita = [ 
    2238.3,
    2019, 
    2029.2,
    2283.5,
    2509.2,
    2608.6,
    2792,
    2826.9,
    2986.1,
    3293,
    3574.1,
    3870,
    4089.6,
    4295.9
]
```

```python
plt.xlabel('Metai')
plt.ylabel('Atlyginimas €')
plt.title('Lietuvos vidutinis atlyginimas (€)')
plt.plot(periods, neto, color='#fff81f', linestyle='--', marker='.', label='Neto')
plt.plot(periods, bruto,  color='b', label='Bruto')
plt.plot(periods, gdp_per_capita,  color='#c9ff05', linewidth=3, label='BVP gyventojui')
plt.legend();
```

Šiuo atveju `plt` gana tvarkingai atvaizduoja mūsų duomenys. Tačiau kartais gali būti kad duomenys esantys ant ašių gali netilpti dėl kokių nors formatavimo ypatybių. Todėl kartais reikės modifikuoti `padding` nustatymus ir tai galima padaryti su metodu `plt.tight_layout()`

```python
plt.xlabel('Metai')
plt.ylabel('Atlyginimas €')
plt.title('Lietuvos vidutinis atlyginimas (€)')
plt.plot(periods, neto, color='#fff81f', linestyle='--', marker='.', label='Neto')
plt.plot(periods, bruto,  color='b', label='Bruto')
plt.plot(periods, gdp_per_capita,  color='#c9ff05', linewidth=3, label='BVP gyventojui')
plt.legend()
plt.tight_layout();
```

Gali pasirodyti kad niekas nepasikeitė, bet priklausomai nuo jūsų monitoriaus ir kitų su rezoliucija susijusiu nustatymų `plt` dabar išnaudoja maksimaliai galima erdvę išdėstyti informacija grafike. Tai pat yra atveju kai pravartu matyti grafiko informacija su pagalbiniu tinklu kuris leidžia optiškai įvertinti duomenys. Mes galime įjungti šį tinklelį su metodu `plt.grid(True)` jam duodami loginės reikšmės argumentą su verte <code>**True**</code>.

```python
plt.xlabel('Metai')
plt.ylabel('Atlyginimas €')
plt.title('Lietuvos vidutinis atlyginimas (€)')
plt.plot(periods, neto, color='#fff81f', linestyle='--', marker='.', label='Neto')
plt.plot(periods, bruto,  color='b', label='Bruto')
plt.plot(periods, gdp_per_capita,  color='#c9ff05', linewidth=3, label='BVP gyventojui')
plt.legend()
plt.tight_layout()
plt.grid(True);
```

### Stulpelinės diagramos

Kadangi naudojome `plt.plot()` metodą, matplotlib mūsų duomenys paverte kreive. Tačiau jeigu šį metodą pakeisime į `plt.bar()` mūsų grafikas matplotlib pakeis į stulpelinės diagramos formą (yra vertikali ir horizontali).

```python
plt.bar(periods, neto, label='Neto')
```

Bet be neto atlyginimo tai pat turime duomenys apie bruto ir bvp vienam gyventojui. Jeigu mes norime šiuos duomenys įtraukti į stulpelinės diagromos grafiką mums tereikia kaip ir su `plt.plot()` metodu tiesiog papildyti savo kodą pipildomomis eilutėmis.<br>
<code>**Pastaba:**</code> reikia atsižvelgti į tai kad stulpelinės diagromos bus piešiamos vieną ant kitos, o ne greta. Todėl jeigu tarkime bvp vienam gyventojui sąrašo reikšmės yra didžiausios jos paprasčiausiai uždengs kitų diagramų duomenys. Todėl eiluškumas yra svarbu!

```python
plt.bar(periods, gdp_per_capita, label='BVP gyventojui')
plt.bar(periods, bruto, label='Bruto')
plt.bar(periods, neto, label='Neto')
plt.legend()
```

Yra ir antras būdas kaip atvaizduoti duomenys stulpelinės diagroms formoje ir išvengti eiliavimo tačiau jis gali pasirodyti šiek tiek komplikuotas. Duomenys galima išdėstyti stulpelines diagramoje vienus greta kitų. Mums reikės šiek tiek paslinkti (offsetting) x ašyje mūsų duomenys. ### Duomenys iš .csv failų

Yra tris populiarus būdai įkelti duomenys iš išorinio failo:
 - [csv.DictReader()](https://docs.python.org/3/library/csv.html#csv.DictReader)
 - [pandas.read_csv()](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html)
 - [numpy.loadtxt()](https://numpy.org/doc/stable/reference/generated/numpy.loadtxt.html)

```python
import csv
```

python [with](https://docs.python.org/3/reference/compound_stmts.html#the-with-statement), [open](https://docs.python.org/3/library/functions.html#open), [next](https://docs.python.org/3/library/functions.html#next) funkcijos

```python
with open('../data_samples/salary_data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    row = next(csv_reader)
    print(row)
    print(row['YearsExperience'])
    print(row['Salary'])
```

```python
years  = []
salary = []
```

```python
with open('../data_samples/salary_data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    for row in csv_reader:
        years.append(row['YearsExperience'])
        salary.append(row['Salary'])
        
```

```python
plt.barh(years, salary)
plt.xticks(rotation=-90);
```
